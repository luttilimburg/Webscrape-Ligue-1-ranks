```python
url = 'https://www.fussballdaten.de/frankreich/tabelle/'  #we define the URL we want to scrap from 

page = requests.get(url)         #this requests the URL to be scraped

soup = BeautifulSoup(page.text, 'lxml') #we define the parser and the page

table = soup.find('table', class_='table lh2') #getting the table

headers = [] #we make a list

for i in table.find_all('th'):  #we iterate through the table to find all headers and append them to our list
    title = i.text
    headers.append(title)

    print(headers)  # we print out the list
    


```

```python
df = pd.DataFrame(columns = headers)  #we use this list as headers for our dataframe
df 
```

```python
for row in table.find_all('tr')[1:]:   #grab the first row and go the td tags and add them into the first row using 4loop
    data = row.find_all('td')  
    row_data = [td.text.strip() for td in data] #get the text of each td
    length = len(df)
    df.loc[length] = row_data    #add the rows to our dataframe
```
